```python
sequence_length =10
x_seq=[]
label=[]
for i in range(len(x)-sequence_length):
  x_seq.append(x[i:i+sequence_length])
  label.append(x[i+sequence_length])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_seq,label,test_size=0.2,random_state=42)
len(x_seq)-(len(x_seq)*0.2),len(label)
```

```python
np.array(x_train).shape,np.array(y_train).shape,np.array(x_test).shape,np.array(y_test).shape
```

```python
model=Sequential()
model.add(layers.SimpleRNN(64,return_sequences=True,input_shape=(10,1)))
model.add(layers.SimpleRNN(32,return_sequences=False))
model.add(layers.Dense(16,activation='tanh'))
model.add(layers.Dense(1,activation='tanh'))
model.compile(optimizer='adam',loss='mse',metrics=['mse','mae'])
model.summary()
```

```python
histroy=model.fit(np.array(x_train),np.array(y_train),validation_split=0.2,epochs=20,verbose=1)
```

```python
y_hat=model.predict(np.array(x_test))
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mean_squared_error(y_test,y_hat)
```

```python
r2_score(y_test,y_hat)
```
